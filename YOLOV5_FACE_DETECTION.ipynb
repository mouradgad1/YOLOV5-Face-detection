## Configurations
setting up the training and validation image sets


```python
import yaml

yaml_path = "/content/yolov5/Face-Detection-1/data.yaml"

with open(yaml_path, 'r') as file:
    data_yaml = yaml.safe_load(file)

data_yaml['train'] = "/content/yolov5/Face-Detection-1/train/images"
data_yaml['val'] = "/content/yolov5/Face-Detection-1/train/images"

with open(yaml_path, 'w') as file:
    yaml.dump(data_yaml, file)

print("Updated data.yaml:", data_yaml)
```
